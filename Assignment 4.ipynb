```python
#15 years: AAPL, GOOG, TSLA, AMZN, MSFT, NFLX, FB.

import pandas as pd
#import datetime from datetime 


# AAPL
df_aapl = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/AAPL?period1=1153094400&period2=1626480000&interval=1d&events=history&includeAdjustedClose=true')

# GOOG
df_goog = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/GOOG?period1=1153094400&period2=1626480000&interval=1d&events=history&includeAdjustedClose=true')

# TSLA
df_tsla = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/TSLA?period1=1153094400&period2=1626480000&interval=1d&events=history&includeAdjustedClose=true')

# AMZN
df_amzn = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/AMZN?period1=1153094400&period2=1626480000&interval=1d&events=history&includeAdjustedClose=true')

# MSFT
df_msft = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/MSFT?period1=1153094400&period2=1626480000&interval=1d&events=history&includeAdjustedClose=true')

# NFLX
df_nflx = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/NFLX?period1=1153094400&period2=1626480000&interval=1d&events=history&includeAdjustedClose=true')

# FB
df_fb = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/FB?period1=1153094400&period2=1626480000&interval=1d&events=history&includeAdjustedClose=true')
```

```python
df_aapl['AAPL Return'] = df_aapl['Adj Close'].pct_change()
df_goog['GOOG Return'] = df_goog['Adj Close'].pct_change()
df_tsla['TSLA Return'] = df_tsla['Adj Close'].pct_change()
df_amzn['AMZN Return'] = df_amzn['Adj Close'].pct_change()
df_msft['MSFT Return'] = df_msft['Adj Close'].pct_change()
df_nflx['NFLX Return'] = df_nflx['Adj Close'].pct_change()
df_fb['FB Return'] = df_fb['Adj Close'].pct_change()
```

```python
df_combined = [
    df_aapl.set_index('Date').rename(columns={'Adj Close':'Adj Close AAPL'}),
    df_goog.set_index('Date').rename(columns={'Adj Close':'Adj Close GOOG'}),
    df_tsla.set_index('Date').rename(columns={'Adj Close':'Adj Close TSLA'}),
    df_amzn.set_index('Date').rename(columns={'Adj Close':'Adj Close AMZN'}),
    df_msft.set_index('Date').rename(columns={'Adj Close':'Adj Close MSFT'}),
    df_nflx.set_index('Date').rename(columns={'Adj Close':'Adj Close NFLX'}),
    df_fb.set_index('Date').rename(columns={'Adj Close':'Adj Close FB'})
]

df_combined = pd.concat(df_combined, join='outer', axis=1)
df_combined
```

```python
#answer to 1aNEED TO EDIT IT TO ONLY the last 1 year
df_past_1_yr = df_combined
df_past_1_yr.index = pd.DatetimeIndex(df_past_1_yr.index)
df_past_1_yr = df_combined.last('365D')
fifteen_yr_return_dict = {}

AAPL_15_yr_return = (df_past_1_yr['Adj Close AAPL'].iloc[-1]/(df_past_1_yr['Adj Close AAPL'].loc[df_past_1_yr['Adj Close AAPL'].first_valid_index()])-1)*100
GOOG_15_yr_return = (df_past_1_yr['Adj Close GOOG'].iloc[-1]/(df_past_1_yr['Adj Close GOOG'].loc[df_past_1_yr['Adj Close GOOG'].first_valid_index()])-1)*100
TSLA_15_yr_return =(df_past_1_yr['Adj Close TSLA'].iloc[-1]/(df_past_1_yr['Adj Close TSLA'].loc[df_past_1_yr['Adj Close TSLA'].first_valid_index()])-1)*100
AMZN_15_yr_return = (df_past_1_yr['Adj Close AMZN'].iloc[-1]/(df_past_1_yr['Adj Close AMZN'].loc[df_past_1_yr['Adj Close AMZN'].first_valid_index()])-1)*100
MSFT_15_yr_return = (df_past_1_yr['Adj Close MSFT'].iloc[-1]/(df_past_1_yr['Adj Close MSFT'].loc[df_past_1_yr['Adj Close MSFT'].first_valid_index()])-1)*100
NFLX_15_yr_return = (df_past_1_yr['Adj Close NFLX'].iloc[-1]/(df_past_1_yr['Adj Close NFLX'].loc[df_past_1_yr['Adj Close NFLX'].first_valid_index()])-1)*100
FB_15_yr_return = (df_past_1_yr['Adj Close FB'].iloc[-1]/(df_past_1_yr['Adj Close FB'].loc[df_past_1_yr['Adj Close FB'].first_valid_index()])-1)*100

fifteen_yr_return_dict = {'AAPL' :AAPL_15_yr_return, 'GOOG':GOOG_15_yr_return, 'TSLA': TSLA_15_yr_return, 'AMZN': AMZN_15_yr_return, 
'MSFT':MSFT_15_yr_return, 'NFLX': NFLX_15_yr_return,'FB':FB_15_yr_return
}

for i in fifteen_yr_return_dict:
    #print(i,Fifteen_yr_return_dict[i]) 
    print(f'{i}: {fifteen_yr_return_dict[i]:.2f}%')

print(f'The stock with the largest return for the past year is {max(fifteen_yr_return_dict, key=fifteen_yr_return_dict.get) } and the return is {max(fifteen_yr_return_dict.values()):.2f}%')
```

```python
#answer to 1e
df_past_yr = df_combined
# Normalize the prices
def normalize_price(x):
    return x / x[0]

df_past_yr.index = pd.DatetimeIndex(df_past_yr.index)
df_past_yr = df_combined.last('365D')
df_past_yr = df_past_yr.apply(normalize_price)
df_past_yr.loc[:, ['Adj Close AAPL', 'Adj Close GOOG', 'Adj Close TSLA','Adj Close AMZN','Adj Close MSFT','Adj Close NFLX','Adj Close FB']].plot(kind='line', figsize=(17, 7), title='Past Year Normalized', grid=True)
```

```python
# SPY since 1995
df_spy_95 = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/SPY?period1=788918400&period2=1626480000&interval=1d&events=history&includeAdjustedClose=true')
df_spy_95['EMA_12'] = df_spy_95['Adj Close'].rolling(12).mean()
df_spy_95['EMA_26'] = df_spy_95['Adj Close'].rolling(26).mean()
df_spy_95['MACD'] =df_spy_95['EMA_26']  - df_spy_95['EMA_12']
df_spy_95.loc[:,['Adj Close','EMA_12','EMA_26']].plot(kind='line', figsize=( 17, 7),title='SPY since 1995', grid=True)
df_spy_95.loc[:,['MACD']].plot(kind='line', color='red', figsize=( 17, 3),title='SPY MACD')
```
