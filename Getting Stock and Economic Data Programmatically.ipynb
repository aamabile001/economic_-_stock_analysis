```python
import pandas_datareader as pdr
import datetime

# I just set the start date to Jan 1, 1900 because it usually gets the 
# earliest available data
start = datetime.datetime(1900, 1, 1)
end = datetime.datetime.now()

# The data we're getting: https://fred.stlouisfed.org/series/GDPC1
# "Real Gross Domestic Product (Quarterly)"
gdp = pdr.DataReader('GDPC1', 'fred', start, end)
gdp
```

```python
gdp.plot()
```

```python
gdp['GDP Change'] = gdp['GDPC1'].pct_change()
gdp
```

```python
gdp[['GDP Change']].plot(kind='line', figsize=(15,8), grid=True)
```

```python
# What do you think this will do?
gdp['GDP Change Annual'] = gdp['GDPC1'].pct_change(4)
gdp
```

```python
gdp[['GDP Change Annual']].plot(kind='line', figsize=(15,8), grid=True)
```

```python
# The data we're getting: https://fred.stlouisfed.org/series/CPIAUCNS
# "Consumer Price Index for All Urban Consumers (Monthly)"
inflation = pdr.DataReader('CPIAUCNS', 'fred', start, end)
inflation['CPIAUCNS'].plot(figsize=(15,8), grid=True)
```

```python
inflation
```

```python
# but that isn't very helpful, we want to change
inflation['CPI Change Annual'] = inflation['CPIAUCNS'].pct_change(12)
inflation['CPI Change Annual'].plot(figsize=(15,8), grid=True)
```

```python
# Also get the 10 year interest rate and amount of public debt
ten_yr = pdr.DataReader('DGS10', 'fred', start, end)
public_debt = pdr.DataReader('GFDEGDQ188S', 'fred', start, end).pct_change(4)
```

```python
# Let's plot these economic variables next to each other
# But do you notice soemthing wrong?
gdp[['GDP Change Annual']].plot(kind='line', figsize=(15,4), grid=True, title='GDP')
inflation[['CPI Change Annual']].plot(kind='line', figsize=(15,4), grid=True, title='inflation')
ten_yr.plot(kind='line', figsize=(15,4), grid=True, title='10 yr')
public_debt.plot(kind='line', figsize=(15,4), grid=True, title='Public Debt Change')
```

```python
# For you to use:
# A function that can download a bunch of data from FRED and merge the Dataframes
def get_fred_data(symbols, start = datetime.datetime(1900, 1, 1), end = datetime.datetime.now()):
    result = None
    for symbol in symbols:
        df = pdr.DataReader(symbol['symbol'], 'fred', start, end)
        df = df.rename(columns={df.columns[0]: symbol['title']})
        if result is None:
            result = df.copy()
        else:
            result = pd.merge_asof(left=result.copy(), right=df, on='DATE')
           
    result = result.set_index('DATE')
    return result
```

```python
symbols = [
    {'symbol': 'A191RL1Q225SBEA', 'title': 'Real GDP Change'},
    {'symbol': 'DGS10', 'title': '10 Year Rate'},
    {'symbol': 'FPCPITOTLZGUSA', 'title': 'CPI'},
    {'symbol': 'GFDEGDQ188S', 'title': 'Public Debt'},
]
fred_data = get_fred_data(symbols)
fred_data
```

```python
fred_data.plot(subplots=True, figsize=(15,10), grid=True)
```

```python
# For you to use:
# A function that can download a bunch of stock tickers and merge the Dataframes
def get_yfinance_symbols(symbols, start = datetime.datetime(1900, 1, 1), end = datetime.datetime.now()):
    result = None
    for symbol in symbols:
        ticker = yf.Ticker(symbol)
        df = ticker.history(period="max")
        df = df.loc[:, ['Close']].rename(columns={'Close': symbol})
        if result is None:
            result = df.copy()
        else:
            result = pd.merge(left=result.copy(), right=df, left_index=True, right_index=True)
           
    return result

df = get_yfinance_symbols(['AAPL', 'MSFT', 'QQQ', 'SPY', 'TSLA', 'AMZN'])
df
```

```python
# Graph it out
df.plot(figsize=(14, 5))
```

```python
# Normalize prices
def normalize(x):
    return x / x[0]

df_norm = df.apply(normalize)
df_norm
```

```python
# Plot
df_norm.plot(kind='line', figsize=(20,8))
```
